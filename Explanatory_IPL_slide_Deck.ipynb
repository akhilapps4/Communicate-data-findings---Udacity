# Investigation overview
In this Investigation, i wanted to look at the features of players that could be used to predict the player performances. The main focus was on the features like 'Batsman', 'Batsman_runs', 'Bowler', 'Player_dismissed' from delivery data and features like 'winner' and 'player_of_match' from matches data


# Dataset Overview
The data sets delivery and matches consists of 15460 ball by ball data and 636 matches with attributes include'Batting_team', 'Bowling_team', 'Season' and 'winner'. Few of these attributes are used for the analysis and visualization
